```python
#change the password with whatever ur password is
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = 'mysecretpassword'
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
#1. Prepare a list of offices sorted by country, state, city.
run("""
    SELECT * FROM offices ORDER BY country ASC, state ASC, city ASC
""")
```

```python
#2. How many employees are there in the company?
run("""
    SELECT COUNT(*) as TOTAL_EMPLOYEE FROM employees
""")
```

```python
#3. What is the total of payments received?
run("""
    SELECT SUM(amount) as TOTAL_PAYMENT_RECEIVED FROM payments
""")
```

```python
#4. List the product lines that contain 'Cars'.
run("""
    SELECT * FROM products WHERE productLine LIKE '%Cars%'
""")
```

```python
#5. Report total payments for October 28, 2004.
run("""
    SELECT SUM(amount) as TOTAL_PAYMENTS_FOR_OCTOBER_28_2004 FROM payments WHERE paymentDate = '2004-10-28'
""")
```

```python
#6. Report those payments greater than $100,000.
run("""
    SELECT * FROM payments WHERE amount > 100000
""")
```

```python
#7. List the products in each product line.
run("""
    SELECT * FROM products ORDER BY productLine ASC, productName ASC
""")
```

```python
#8. How many products in each product line?
run("""
    SELECT productLine, COUNT(*) as TOTAL FROM products GROUP BY productLine ORDER BY productLine ASC
""")
```

```python
#9. What is the minimum payment received?
run("""
    SELECT  MIN(amount) as MINIMUM_PAYMENT_RECEIVED FROM payments 
""")
```

```python
#10. List all payments greater than twice the average payment.
run("""
    SELECT * FROM payments WHERE amount >  2 * (SELECT AVG(amount) FROM payments)
""")
```

```python
#11. What is the average percentage markup of the MSRP on buyPrice?
run("""
    SELECT AVG(((MSRP - buyPrice) / MSRP) * 100) as AVERAGE_PERCENTAGE_MARKUP_OF_THE_MSRP_ON_BUYPRICE FROM products
""")

```

```python
#12. How many distinct products does ClassicModels sell?
run("""
    SELECT productCode, COUNT(*) as PRODUCT_ORDERED FROM orderdetails WHERE productCode IN ((SELECT productCode FROM products WHERE productLine LIKE '%Classic%' )) GROUP BY productCode
""")
```

```python
#13. Report the name and city of customers who don't have sales representatives
run("""
    SELECT customerName, city FROM customers WHERE salesRepEmployeeNumber IS NULL ORDER BY customerName ASC
""")
```

```python
#14. What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.
run("""
    SELECT CONCAT(firstName, ' ', lastName) as NAME, jobTitle FROM employees WHERE jobTitle LIKE '%VP%' OR jobTitle LIKE '%Manager%'
""")
```

```python
#15. Which orders have a value greater than $5,000?
run("""
    SELECT o.*, SUM(od.priceEach * od.quantityOrdered) as TOTAL 
        FROM orderdetails o 
        INNER JOIN orderdetails od ON o.orderNumber = od.orderNumber
        GROUP BY o.orderNumber 
        HAVING TOTAL > 5000 
""")
```
