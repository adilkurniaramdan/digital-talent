```python
import pandas as pd
data_faktur_masukan = pd.read_csv('FM_PELINDO_2018.csv', delimiter = ';', encoding = "ISO-8859-1")
```

```python
#Deskripsi Data
data_faktur_masukan.iloc[:,1:14]
```

```python
#2. Statistik tentang data
# mean JUMLAH_DPP 
data_faktur_masukan.groupby('NPWP').mean().iloc[:,5:8]
```

```python
#2. Statistik tentang data
#data pemebelian perbulan
data_faktur_masukan_per_bulan = data_faktur_masukan.groupby('MASA_PAJAK').sum().iloc[:,5:8]
data_faktur_masukan_per_bulan
```

```python
ax = data_faktur_masukan.groupby('MASA_PAJAK').sum()['JUMLAH_DPP']
ax = ax.plot.bar(stacked=True);
ax.set_xlabel("Bulan", fontsize=12)
ax.set_ylabel("JUMLAH_DPP", fontsize=12)
```

   + data pemasukan dari faktur dikelompokan per bulan
- selain dari itu juga menampilkan bar chart untuk menggambarkan penjualan perbulan
