```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = 'mysecretpassword'
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
# One to many relationship
#1.Report the account representative for each customer.
run("""
    SELECT c.* FROM customers c LEFT JOIN employees e 
        ON c.salesRepEmployeeNumber = e.employeeNumber
""")
```

```python
#2.Report total payments for Atelier graphique.
run("""
    SELECT * FROM customers c INNER JOIN payments USING(customerNumber) WHERE c.customerName = 'Atelier graphique' 
""")
```

```python
# One to many relationship
#3.Report the total payments by date
run("""
    SELECT p.paymentDate, SUM(p.amount) as TOTAL_PAYMENT_BY_DATE FROM payments p 
        GROUP BY p.paymentDate 
        ORDER BY p.paymentDate ASC
""")
```

```python
# One to many relationship
#4. Report the products that have not been sold.
run("""
    SELECT p.* FROM products p 
        LEFT JOIN orderdetails  od ON p.productCode = od.productCode
        WHERE od.productCode IS NULL
        
""")
```

```python
# One to many relationship
#5. List the amount paid by each customer.
run("""
    SELECT c.customerNumber, c.customerName, SUM(p.amount) as TOTAL_PAID FROM customers c 
        INNER JOIN payments p USING (customerNumber) 
        GROUP BY c.customerNumber
""")
```

```python
# One to many relationship
#6. How many orders have been placed by Herkku Gifts?
run("""
    SELECT c.customerName, o.*  FROM customers c 
        INNER JOIN orders o USING(customerNumber)
        WHERE c.customerName = 'Herkku Gifts' 
""")
```

```python
# One to many relationship
#7. Who are the employees in Boston?
run("""
    SELECT * FROM employees e 
        INNER JOIN offices o USING(officeCode)
        WHERE o.city = 'Boston'
""")
```

```python
# One to many relationship
#8. Report those payments greater than $100,000. Sort the report so the customer who made the highest payment appears first.
run("""
    SELECT 
        c.customerNumber, 
        c.customerName, 
        p.amount
    FROM customers c
        INNER JOIN payments p ON c.customerNumber = p.customerNumber AND p.amount > 100000
        ORDER BY p.amount DESC
""")
```

```python
# One to many relationship
#9. List the value of 'On Hold' orders.
run("""
    SELECT 
        o.orderNumber,
        o.orderDate,
        o.status,
        SUM(od.quantityOrdered * od.priceEach) as TOTAL
    FROM orders o
        INNER JOIN orderdetails od ON o.orderNumber = od.orderNumber AND o.status = 'On Hold'
        GROUP BY o.orderNumber
""")
```

```python
# One to many relationship
#10. Report the number of orders 'On Hold' for each customer.
run("""
    SELECT 
        c.customerNumber,
        c.customerName,
        o.status,
        COUNT(1) as NUMBER_ON_HOLD_ORDER_CUSTOMER
    FROM customers c 
        INNER JOIN orders o ON c.customerNumber = o.customerNumber
        INNER JOIN orderdetails od ON o.orderNumber = od.orderNumber AND o.status = 'On Hold'
        GROUP BY c.customerNumber 
        ORDER BY c.customerNumber ASC
""")
```

```python
#Many to many relationship
#1. List products sold by order date
run("""
    SELECT 
        o.orderDate,
        o.orderNumber,
        p.productCode,
        p.productName,
        od.quantityOrdered,
        od.priceEach,
        (od.quantityOrdered * od.priceEach) as total
    FROM products p 
    INNER JOIN orderdetails od ON p.productCode = od.productCode
    INNER JOIN orders o ON od.orderNumber = o.orderNumber
    ORDER BY o.orderDate ASC
""")
```

```python
#Many to many relationship
#2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.
run("""
    SELECT 
        o.orderDate,
        o.orderNumber,
        p.productCode,
        p.productName,
        od.quantityOrdered,
        od.priceEach,
        (od.quantityOrdered * od.priceEach) as total
    FROM products p 
    INNER JOIN orderdetails od ON p.productCode = od.productCode
    INNER JOIN orders o ON od.orderNumber = o.orderNumber
    WHERE p.productName = '1940 Ford Pickup Truck'
    ORDER BY o.orderDate DESC
""")
```

```python
#Many to many relationship
#3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?
run("""
    SELECT 
        c.customerNumber,
        c.customerName,
        o.orderNumber,
        o.orderDate,
        SUM(od.quantityOrdered * od.priceEach) as total
    FROM customers c 
    INNER JOIN orders o ON o.customerNumber = c.customerNumber
    INNER JOIN orderdetails od ON od.orderNumber = o.orderNumber 
    GROUP BY o.orderNumber
    HAVING total > 25000
""")
```

```python
#Many to many relationship
#4. Are there any products that appear on all orders?
run("""
    SELECT 
        p1.productCode,
        p1.productName
    FROM products p1
    HAVING (SELECT COUNT(1)
                FROM products p
                INNER JOIN orderdetails od ON od.productCode = p.productCode
                INNER JOIN orders o ON o.orderNumber = od.orderNumber
                WHERE p.productCode = p1.productCode
        ) = (SELECT COUNT(1) FROM orders)
        
""")
```

```python
#Many to many relationship
#5. List the names of products sold at less than 80% of the MSRP.
run("""
    SELECT 
        DISTINCT
        p1.productCode,
        p1.productName,
        p1.MSRP,
        (((p1.MSRP - p1.buyPrice) / p1.MSRP) * 100) as PERCENTAGE_OF_MSRP
    FROM products p1 
    INNER JOIN orderdetails od USING(productCode)
    INNER JOIN orders o USING(orderNumber)
    HAVING PERCENTAGE_OF_MSRP < 80
        
""")
```

```python
#Many to many relationship
#6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)
run("""
    SELECT 
        o.orderNumber,
        p.productCode,
        p.productName,
        p.buyPrice,
        od.priceEach,
        (od.priceEach - p.buyPrice) as MARK_UP
    FROM products p
    INNER JOIN orderdetails od ON od.productCode = p.productCode
    INNER JOIN orders o ON o.orderNumber = od.orderNumber 
    WHERE od.priceEach >= (2 * p.buyPrice) 

        
""")
```

```python
#Many to many relationship
#7. List the products ordered on a Monday.

run("""
    SELECT 
        o.orderNumber,
        o.orderDate,
        DAYNAME(o.orderDate) as day_of_week,
        p.productCode,
        p.productName
    FROM products p
    INNER JOIN orderdetails od ON od.productCode = p.productCode
    INNER JOIN orders o ON o.orderNumber = od.orderNumber 
    WHERE DAYNAME(o.orderDate) = 'Monday'
    ORDER BY o.orderDate ASC

        
""")
```

```python
#Many to many relationship
#8. What is the quantity on hand for products listed on 'On Hold' orders

run("""
    SELECT 
        p.productCode,
        p.productName,
        p.quantityInStock,
        o.status,
        SUM(od.quantityOrdered) as DEMAND
    FROM products p
    INNER JOIN orderdetails od ON od.productCode = p.productCode
    INNER JOIN orders o ON o.orderNumber = od.orderNumber AND o.status = 'On Hold'
    GROUP BY p.productCode
        
""")
```
